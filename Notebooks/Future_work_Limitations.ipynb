I did not use SHAP due to time and computational cost. SHAP gives detailed explanations but is slow on large tabular datasets with many encoded features. Given the scope of this assessment, I prioritized model training and evaluation. SHAP remains a good option for future work once more time is available.

### **Limitations:**

One main limitation is the class imbalance. Techniques like SMOTE or undersampling can sometimes help, but they also have drawbacks: SMOTE may generate synthetic samples that add noise, and undersampling risks removing useful information. In addition, since the final model is tree-based, SMOTE can distort the distribution and lead to unrealistic samples. For categorical features, oversampling can also create combinations of categories that do not exist in the real data. For these reasons, I chose not to apply SMOTE or similar resampling methods in this project.
